```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/vaoni/penguins_size.csv')
```

```python
df = df.dropna()
```

```python
X = pd.get_dummies(df.drop('species',axis = 1),drop_first=True)
```

```python
y =df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=10,random_state=101)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report
```

```python
plot_confusion_matrix(rf,X_test,y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
X = dff.drop('Class',axis = 1)
```

```python
y = dff['Class']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
param_grid = {'n_estimators':[64,100,128,200],'max_features':[2,3,4],'bootstrap':[True,False],'oob_score':[True,False]}
```

```python
rfc = RandomForestClassifier()
```

```python
grid_mod = GridSearchCV(rfc,param_grid)
```

```python
grid_mod.fit(X_train,y_train)
```

```python
grid_mod.best_params_
```

```python
preds = grid_mod.predict(X_test)
```

```python
plot_confusion_matrix(grid_mod,X_test,y_test)
```

```python
print(classification_report(y_test,preds))
```
